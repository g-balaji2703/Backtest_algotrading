```python
def MACD(data,ma_fast,ma_slow,signal):
    df = data.copy()
    df['ma_fast'] = df['Close'].ewm(span=ma_fast,min_periods=ma_fast).mean()
    df['ma_slow'] = df['Close'].ewm(span=ma_slow,min_periods=ma_slow).mean()
    df['MACD'] = df['ma_fast'] - df['ma_slow']
    df['Signal'] = df['MACD'].ewm(span=signal,min_periods=signal).mean()
    df['Histogram'] = df['MACD'] - df['Signal']
    return df
```
